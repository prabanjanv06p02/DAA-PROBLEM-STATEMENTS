```python
import random
import time


def interpolation_search(arr, target):
    """
    Interpolation Search for sorted floating-point numbers.
    Returns:
        index: position of target (-1 if not found)
        comparisons: number of comparisons performed
    """
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= target <= arr[high]:
        comparisons += 2  # range checks

        if low == high:
            comparisons += 1
            if arr[low] == target:
                return low, comparisons
            return -1, comparisons

        # Estimate the position using interpolation
        pos = low + int(
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        comparisons += 1
        if arr[pos] == target:
            return pos, comparisons

        comparisons += 1
        if arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


def test_interpolation_search():
    dataset_sizes = [10000, 50000, 100000]
    trials = 1000

    print("=" * 60)
    print("Interpolation Search on Floating-Point Numbers")
    print("Range: 0.0 to 1000.0 (Uniform Distribution)")
    print("=" * 60)

    for size in dataset_sizes:
        # Generate sorted floating-point dataset
        data = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

        total_comparisons = 0
        total_time = 0

        for _ in range(trials):
            # Select a target guaranteed to exist
            target = data[random.randint(0, size - 1)]

            start = time.perf_counter()
            index, comparisons = interpolation_search(data, target)
            end = time.perf_counter()

            total_comparisons += comparisons
            total_time += (end - start)

        avg_comparisons = total_comparisons / trials
        avg_time = (total_time / trials) * 1_000_000  # microseconds

        print(f"\nDataset Size: {size:,}")
        print(f"Average Comparisons: {avg_comparisons:.2f}")
        print(f"Average Search Time: {avg_time:.2f} microseconds")

    print("\nCompleted.")


if __name__ == "__main__":
    test_interpolation_search()
```
